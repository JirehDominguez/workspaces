# Kruskal's Algorithm

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])

edges_with_weights = [
    ("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15),
    ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35),
    ("F", "G", 50), ("G", "E", 45)
]

G.add_weighted_edges_from(edges_with_weights)

mst = nx.minimum_spanning_tree(G, algorithm='kruskal')

plt.figure(figsize=(12, 8))
pos = nx.planar_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800, edge_color='black')
nx.draw_networkx_edge_labels(G, pos, edge_labels={edge[:2]: edge[2] for edge in edges_with_weights}, font_color='red')
plt.title("Original Graph with Edge Weights")
plt.show()

plt.figure(figsize=(12, 8))
nx.draw(mst, pos, with_labels=True, node_color='lightgreen', node_size=800, edge_color='blue')
nx.draw_networkx_edge_labels(mst, pos, edge_labels={(u, v): d['weight'] for u, v, d in mst.edges(data=True)}, font_color='red')
plt.title("Minimum Spanning Tree using Kruskal's Algorithm")
plt.show()
```

# Prim's Algorithm

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])

edges_with_weights = [
    ("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15),
    ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35),
    ("F", "G", 50), ("G", "E", 45)
]

G.add_weighted_edges_from(edges_with_weights)

mst_prim = nx.minimum_spanning_tree(G, algorithm='prim')

plt.figure(figsize=(12, 8))
pos = nx.planar_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800, edge_color='black')
nx.draw_networkx_edge_labels(G, pos, edge_labels={edge[:2]: edge[2] for edge in edges_with_weights}, font_color='red')
plt.title("Original Graph with Edge Weights")
plt.show()

plt.figure(figsize=(12, 8))
nx.draw(mst_prim, pos, with_labels=True, node_color='lightgreen', node_size=800, edge_color='blue')
nx.draw_networkx_edge_labels(mst_prim, pos, edge_labels={(u, v): d['weight'] for u, v, d in mst_prim.edges(data=True)}, font_color='red')
plt.title("Minimum Spanning Tree using Prim's Algorithm")
plt.show()
```

# Dijkstra's Algorithm

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])

edges_with_weights = [
    ("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15),
    ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35),
    ("F", "G", 50), ("G", "E", 45)
]

G.add_weighted_edges_from(edges_with_weights)

shortest_paths = nx.single_source_dijkstra_path(G, source="A")

shortest_path_edges = [(source, target) for source in shortest_paths for target in shortest_paths[source][1:]]
G_shortest_paths = G.edge_subgraph(shortest_path_edges)

plt.figure(figsize=(12, 8))
pos = nx.planar_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=800, edge_color='black')
nx.draw_networkx_edge_labels(G, pos, edge_labels={edge[:2]: edge[2] for edge in edges_with_weights}, font_color='red')
plt.title("Original Graph with Edge Weights")
plt.show()

plt.figure(figsize=(12, 8))
nx.draw(G_shortest_paths, pos, with_labels=True, node_color='lightgreen', node_size=800, edge_color='blue')
plt.title("Shortest Paths from Node A using Dijkstra's Algorithm")
plt.show()

```
