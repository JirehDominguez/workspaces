## Task 1: Question 1

### Derivation of Expected Value:

Given:
- Experiment with $n$ identical trials.
- Each trial has $k$ possible outcomes with probabilities $p_i$.
- Random variables of interest are $Y_i$ representing the number of trials resulting in outcome $i$.

We aim to find the expected value $E[Y]$.

#### Step 1: Definition of Expected Value

The expected value is the sum of each outcome multiplied by its probability:
$E[Y] = \sum_{i=1}^{k} y_i \cdot P(Y_i = y_i)$

#### Step 2: Substitute Joint Probability Distribution

Given the joint probability distribution:
$P(Y_1 = y_1, Y_2 = y_2, ..., Y_k = y_k) = \frac{n!}{y_1!y_2!...y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot ... \cdot p_k^{y_k}$

#### Step 3: Expand the Sum

$E[Y] = \sum_{y_1}\sum_{y_2}...\sum_{y_k} \frac{n!}{y_1!y_2!...y_k!} \cdot (y_1 \cdot p_1) \cdot (y_2 \cdot p_2) \cdot ... \cdot (y_k \cdot p_k)$

#### Step 4: Rearrange Terms

$E[Y] = \frac{n!}{y_1!y_2!...y_k!} \cdot \left(\sum_{y_1} y_1 \cdot p_1\right) \cdot \left(\sum_{y_2} y_2 \cdot p_2\right) \cdot ... \cdot \left(\sum_{y_k} y_k \cdot p_k\right)$

#### Step 5: Recognize Individual Sums as Expected Values of Binomial Distributions

$E[Y_i] = n \cdot p_i$

#### Step 6: Substitute Expected Values

$E[Y] = \frac{n!}{y_1!y_2!...y_k!} \cdot (n \cdot p_1) \cdot (n \cdot p_2) \cdot ... \cdot (n \cdot p_k)$

#### Step 7: Simplify

$E[Y] = n \cdot (p_1 \cdot p_2 \cdot ... \cdot p_k)$

#### Step 8: Recall Probability Summation

$p_1 + p_2 + ... + p_k = 1$

So,

$E[Y] = n \cdot 1$

$E[Y] = n$

#### Conclusion:

The expected value $E[Y]$ equals $n$, the total number of trials.




## Task 1: Question 2

### Given Data:
- Sample size $n = 500$
- Proportion of adults in each age group:

\[
\begin{array}{c|c}
\text{Age} & \text{Proportion} \\
\hline
18-24 & 0.18 \\
25-34 & 0.23 \\
35-44 & 0.16 \\
45-64 & 0.27 \\
65-100 & 0.16 \\
\end{array}
\]

We are asked to find the probability that the sample contains 100 people between 18 and 24, 200 between 25 and 34, 200 between 45 and 64, and the expected value for obtaining a person aged 65 and above.

### Solution:

#### Probability Calculation:
To find the probability, we use the multinomial distribution formula:

$P(Y_1 = 100, Y_2 = 200, Y_3 = 0, Y_4 = 200, Y_5 = 0) = \frac{n!}{100! \cdot 200! \cdot 0! \cdot 200! \cdot 0!} \cdot (0.18)^{100} \cdot (0.23)^{200} \cdot (0.16)^0 \cdot (0.27)^{200} \cdot (0.16)^0$

$P(Y_1 = 100, Y_2 = 200, Y_3 = 0, Y_4 = 200, Y_5 = 0) = \frac{500!}{100! \cdot 200! \cdot 200!} \cdot (0.18)^{100} \cdot (0.23)^{200} \cdot (0.27)^{200}$

#### Expected Value Calculation:
The expected value for obtaining a person aged 65 and above is:

$E[Y_{65+}] = n \cdot p_{65-100} = 500 \cdot 0.16 = 80$

### Conclusion:
The probability that the sample contains 100 people between 18 and 24, 200 between 25 and 34, and 200 between 45 and 64 is calculated using the multinomial distribution. The expected value to obtain a person aged 65 and above is 80.


## Task 2: Question 1

### Derivation of Probability Distribution:

Given:
- Experiment with $n$ identical trials.
- Each trial has two outcomes: success ($S$) or failure ($F$).
- Probability of success on a single trial is $p$, failure probability is $q = (1 - p)$.
- Random variable of interest is $Y$, the number of successes observed during the $n$ trials.

We aim to derive the probability distribution of $Y$.

#### Solution:

The probability distribution follows the binomial distribution, given by:

$P(Y = y) = \binom{n}{y} \cdot p^y \cdot q^{n-y}$

Where:
- $\binom{n}{y}$ is the binomial coefficient, representing the number of ways to choose $y$ successes out of $n$ trials.
- $p^y$ represents the probability of having $y$ successes.
- $q^{n-y}$ represents the probability of having $(n - y)$ failures.

### Conclusion:
The probability distribution of $Y$ is given by the binomial distribution formula.

- Probability of success (recovery) \( p = 0.30 \)
- Number of trials \( n = 10 \)
- Number of successes (recoveries) \( y = 9 \)

We aim to find the probability that at least nine out of ten people receiving the medication will recover.

#### Solution:

Using the binomial probability mass function:

$P(Y \geq 9) = P(Y = 9) + P(Y = 10)$

$P(Y \geq 9) = \binom{10}{9} \cdot (0.30)^9 \cdot (1 - 0.30)^{10-9} + \binom{10}{10} \cdot (1 - 0.30)^{10-10}$

$P(Y \geq 9) = 10 \cdot (0.30)^9 \cdot (0.70)^1 + 1 \cdot (0.30)^{10} \cdot (0.70)^0$

$P(Y \geq 9) = 0.000137781 + 0.0000059049$

$P(Y \geq 9) \approx 0.0001436869 \text{ or } 0.014\%$

### Conclusion:

The probability that at least nine out of ten people receiving the medication will recover is approximately \( 0.014\% \).


## Task 2: Question 3

### Probability Calculation:

Given:
- Probability of success (a fuse being non-defective) \( p = 0.95 \)
- Number of trials \( n = 5 \)

We aim to find the probability of observing at least one defective fuse when a sample of 5 fuses is tested.

#### Solution:

Using the complement rule, we find the probability of observing at least one defective fuse by subtracting the probability of observing zero defective fuses from 1.

$P(\text{At least one defective}) = 1 - P(\text{No defective})$

$P(\text{At least one defective}) = 1 - P(Y = 0)$

Using the binomial probability mass function:

$P(Y = 0) = \binom{5}{0} \cdot (0.05)^0 \cdot (1 - 0.05)^{5-0}$

$P(Y = 0) = (0.95)^5$

$P(\text{At least one defective}) = 1 - (0.95)^5$

$P(\text{At least one defective}) \approx 1 - 0.7738$

$P(\text{At least one defective}) \approx 0.2262$

### Conclusion:

The probability of observing at least one defective fuse in a sample of 5 fuses is approximately \( 0.2262 \) or \( 22.62\% \).

## Task 3: Question 1

### Derivation of Expected Value:

Given:
- Probability distribution $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$

We aim to find a general formula for the expected value of this distribution.

#### Solution:

The expected value $E[Y]$ of a discrete random variable $Y$ with probability mass function $p(y)$ is given by:

$E[Y] = \sum_{y=0}^{\infty} y \cdot p(y)$

Substituting the given probability distribution:

$E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$

Using the definition of the exponential function:

$e^{\lambda} = \sum_{y=0}^{\infty} \frac{\lambda^y}{y!}$

We recognize this as the Taylor series expansion of $e^{\lambda}$, which converges to $e^{\lambda}$.

So, 

$E[Y] = \lambda \cdot e^{-\lambda} \cdot e^{\lambda}$

$E[Y] = \lambda$

### Conclusion:
The general formula for the expected value $E[Y]$ of the given distribution is $\lambda$.
