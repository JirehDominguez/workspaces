### Question 1:
**Proof by Contradiction:**

1. **Assumption:** Let's assume for the sake of contradiction that Euler's number $e$ is rational, expressed as $e = \frac{p}{q}$ where $p$ and $q$ are integers and $q \neq 0$.

2. **Recall the Definition of $e$:** Euler's number $e$ is defined as the sum of the series $\sum_{k=0}^\infty \frac{1}{k!}$.

3. **Partial Sum of the Series:** Consider the partial sum $S_n$ of the series up to $n$ terms: 
\begin{equation*}
S_n = \frac{1}{0!} + \frac{1}{1!} + \frac{1}{2!} + \ldots + \frac{1}{n!}
\end{equation*}

4. **Manipulation of Partial Sum:** We can express $q! \cdot S_n$ as: 
\begin{equation*}
q! \cdot S_n = q! \cdot \left(\frac{1}{0!} + \frac{1}{1!} + \frac{1}{2!} + \ldots + \frac{1}{n!}\right)
\end{equation*}

5. **Expression Expansion:** This expands to: 
\begin{equation*}
q! \cdot S_n = q! + \frac{q!}{1!} + \frac{q!}{2!} + \ldots + \frac{q!}{n!}
\end{equation*}

6. **Contradiction:** However, $q! \cdot S_n$ should be very close to $p$, but $p - q! \cdot S_n$ results in a non-zero positive integer, contradicting the assumption that $e = \frac{p}{q}$.

7. **Conclusion:** Thus, Euler's number $e$ cannot be rational, implying $e \notin \mathbb{Q}$.

### Question 2:
**Minkowski's Inequality Proof:**
1. Use of Holder's Inequality: Start with Holder's Inequality:
\begin{equation*}
\sum_{k=1}^n |a_k b_k| \leq \left( \sum_{k=1}^n |a_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^n |b_k|^q \right)^{\frac{1}{q}}
\end{equation*}

2. Definition of Variables: Define $a_k = |x_k|^p$ and $b_k = |y_k|^p$.

3. Application of Holder's Inequality: Apply Holder's Inequality with the new definitions:
\begin{equation*}
\sum_{k=1}^n |x_k y_k|^p \leq \left( \sum_{k=1}^n |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^n |y_k|^q \right)^{\frac{1}{q}}
\end{equation*}

4. Take the $p$-th Root: Take the $p$-th root of both sides:
\begin{equation*}
\left( \sum_{k=1}^n |x_k y_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^n |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^n |y_k|^p \right)^{\frac{1}{q}}
\end{equation*}

5. Use of Relationship between $p$ and $q$: Recognize that $\frac{p-1}{p} = 1 - \frac{1}{p}$.

6. Apply Holder's Inequality Again: Apply Holder's Inequality to eliminate the $\frac{p-1}{p}$ term:
\begin{equation*}
\left( \sum_{k=1}^n |y_k|^p \right)^{\frac{p-1}{p}} \leq n^{\frac{p-1}{p}} \left( \sum_{k=1}^n |y_k|^p \right)^{\frac{1}{p}}
\end{equation*}

7. Substitute Back into the Inequality: Substitute this equation back into the inequality:
\begin{equation*}
\left( \sum_{k=1}^n |x_k + y_k|^p \right)^{\frac{1}{p}} \leq n^{\frac{1}{p}} \left( \sum_{k=1}^n |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^n |y_k|^p \right)^{\frac{1}{p}}
\end{equation*}


### Question 3:
**Triangle Inequality Proof:**

1. **Absolute Value Definition:** We define the absolute value of a real number $x$ as: 
\begin{equation*}
|x| = \begin{cases} x, & \text{if } x \geq 0 \\ -x, & \text{if } x < 0 \end{cases}
\end{equation*}

2. **Case Analysis:** We consider four cases based on the signs of $x$ and $y$.

3. **Case 1:** When $x \geq 0$ and $y \geq 0$, it directly follows that: 
\begin{equation*}
|x + y| = x + y \leq |x| + |y|
\end{equation*}

4. **Case 2:** Similarly, when $x < 0$ and $y < 0$, we have: 
\begin{equation*}
|x + y| = -(x + y) = -x - y \leq |x| + |y|
\end{equation*}

5. **Case 3:** For $x \geq 0$ and $y < 0$, we show that: 
\begin{equation*}
|x + y| \leq |x| - y
\end{equation*}

6. **Case 4:** Similarly, when $x < 0$ and $y \geq 0$, we have: 
\begin{equation*}
|x + y| \leq -x + |y|
\end{equation*}

7. **Conclusion:** In all cases, we establish: 
\begin{equation*}
|x + y| \leq |x| + |y|
\end{equation*}, thus proving the triangle inequality.

### Question 4:
**Sedrakayan's Lemma Proof:**

1. Use of Cauchy-Schwarz Inequality: We begin by invoking the Cauchy-Schwarz Inequality:
\begin{equation*}
\left(\sum_{i=1}^n u_i v_i\right)^2 \leq \left(\sum_{i=1}^n u_i^2\right) \left(\sum_{i=1}^n v_i^2\right)
\end{equation*}

2. Definition of Variables: Define $\mathbf{u} = (u_1, u_2, ..., u_n)$ and $\mathbf{v} = (\sqrt{v_1}, \sqrt{v_2}, ..., \sqrt{v_n})$.

3. Application of Cauchy-Schwarz: Apply Cauchy-Schwarz Inequality to the vectors $\mathbf{u}$ and $\mathbf{v}$:
\begin{equation*}
\left(\sum_{i=1}^n u_i \sqrt{v_i}\right)^2 \leq \left(\sum_{i=1}^n u_i^2\right) \left(\sum_{i=1}^n v_i\right)
\end{equation*}

4. Rearrangement: Rearrange the inequality:
\begin{equation*}
\frac{\left(\sum_{i=1}^n u_i\right)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
\end{equation*}

5. Conclusion: Sedrakayan's Lemma is established, affirming that for any positive real numbers $u_i$ and $v_i$:
\begin{equation*}
\frac{\left(\sum_{i=1}^n u_i\right)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i}
\end{equation*}
